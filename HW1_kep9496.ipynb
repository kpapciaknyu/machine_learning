**Question 3 (40 points)**: Write a class about your chosen topic (similar complexity to the student example from lecture). This class should have 3 attributes and 2 methods. Provide 2 instance examples. I mostly want to see that you know how to set up a class and that you understand what the basic principles are.  

Save this notebook as HW1_KerberosID.ipynb and upload to brightspace

```python
class Ticket:
  def __init__(self, artist_name= None, ticket_price= None, ticket_sale_price=None , seat_number= None):
    self.artist_name = artist_name
    self.ticket_price = ticket_price
    self.ticket_sale_price = ticket_sale_price
    self.seat_number = seat_number

  def full_ticket_info(self):
    return f"Artist Name: {self.artist_name}, Ticket Price: {self.ticket_price}, Ticket Sale Price: {self.ticket_sale_price}, Seat Number: {self.seat_number}"

  def percent_off_value(self):
    return (float(self.ticket_price) - float(self.ticket_sale_price)) / float(self.ticket_price)

ticket1 = Ticket(artist_name="Charli XCX", ticket_price="100", ticket_sale_price="100", seat_number="14A")
ticket2 = Ticket(artist_name="Troye Sivan", ticket_price="80", ticket_sale_price="60", seat_number="33F")

print("Artist Name:",ticket1.artist_name)
print(ticket1.percent_off_value())
print("Artist Name:",ticket2.artist_name)
print(ticket2.percent_off_value())
```
